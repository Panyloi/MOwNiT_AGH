```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def v_point():
    s = np.random.uniform(0, 2*np.pi)
    t = np.random.uniform(0, np.pi) 
    return np.array([np.cos(s)*np.sin(t), np.sin(s)*np.sin(t), np.cos(t)])
```

```python

def plot_3D_points(points):
    fig = plt.figure()
    ax = fig.add_subplot(projection = '3d')
    for point in points:
        ax.scatter(point[0], point[1], point[2], color = "black")
    # ax.plot_wireframe(points[0][:], points[1][:], points[2][:])
    plt.show()
```

```python
n = 200
v_points = [v_point() for _ in range(n)]
plot_3D_points(v_points)
```

```python
def generate_matrix():
    return np.random.uniform(0, 10**5, (3, 3))
```

```python
print(v_points)
```

```python
x = generate_matrix()
v_points_cp = v_points @ x
```

```python
plot_3D_points(v_points_cp)
```
