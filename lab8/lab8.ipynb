```python
def power_method(A,  iterations=10000, epsilon=1e-9):
    n = len(A)
    r = np.ones(n)
    i = 0
    while i < iterations:
        i+=1
        r_0 = A @ r
        norm = np.linalg.norm(r_0, ord = 1)
        r_norm = r_0/ norm
        if np.linalg.norm(r_norm-r, ord = 1) < epsilon:
            r = r_norm
            break
        r = r_norm

    return norm, (r/np.linalg.norm(r))
```
